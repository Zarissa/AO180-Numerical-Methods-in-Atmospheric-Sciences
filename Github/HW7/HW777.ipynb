```python
#*****************NOTES:********************#




# from numpy import math
import numpy as np
get_ipython().run_line_magic('matplotlib', 'inline')
import matplotlib.pyplot as plt
import matplotlib
import math
import pandas as pd
np.core
from sys import maxsize
from numpy import set_printoptions
set_printoptions(threshold=maxsize)

print("hi")

#*****************Constants********************#

Lx = 2000
Nx = 201
Lz = 2000
Nz = 201
dx = Lx/(Nx-1)
dz = Lz/(Nz-1)
print("dx = ",dx)
print("dz  = ",dz)

g = 9.81
r_0 = 250

d_theta = 0.5
theta_0 = 300

#r = 500.0
tol = 10**-7

#a = 180*1000
#b = 600*1000

x_0 = (Lx/2) 
z_0 = 260


Ttot = 620 #SECONDS
dt = 0.5
Nt = int(Ttot/dt)

U = math.sqrt(2*r_0*g*d_theta/theta_0)
print("U = ",U)
Re = 1500
Kx = 2*r_0*U/Re
Kz = Kx
print("Kx =",Kx)


#****************************************#
#*************Functions******************#
#****************************************#

def forceEqual(array): #left right symmetry
    finarr = np.transpose(array)
    i = 0
    while (i != 52):
        array[i] = array[102-i]
        i+=1
    return np.transpose(finarr)

def twoDSum (array):
    total = 0
    array = np.copy(np.absolute(array))
    for i in array:
        total += math.fsum(i)
    return total


def U_func (PHI):
    dz = 2000
    U = np.zeros((Nx+2, Nz+2), dtype = float)
    for i in range(1,Nx+1):
        for j in range(1,Nz+1):
            U[i][j] = (PHI[i][j-1]-PHI[i][j+1])/(2*dz)
    return (U)
    
def V_func (PHI):
    dx = 2000
    V = np.zeros((Nx+2, Nz+2), dtype = float)
    for i in range(2,Nx):
        for j in range(2,Nz):
            V[i][j] = (PHI[i+1][j]-PHI[i-1][j])/(2*dx)
    return (V)


def J_1_func (PHI, omega): #return standard size Nx+2 x Nz+2 array
    J_1 = np.zeros((Nx+2, Nz+2), dtype = float)
    for i in range(2, Nx):
        for j in range(2,Nz):
            J_1[i][j] = (1/(2*dx))*(PHI[i+1][j]-PHI[i-1][j])*(1/(2*dz))*(omega[i][j+1] - omega[i][j-1]) - (1/(2*dz))*(PHI[i][j+1] - PHI[i][j-1])*(1/(2*dx))*(omega[i+1][j]-omega[i-1][j])
    return (J_1)

def J_2_func (PHI, omega):
    J_2 = np.zeros((Nx+2, Nz+2), dtype = float)
    for i in range(2,Nx):
        for j in range(2,Nz):
            J_2[i][j] = (1/(2*dx))*(PHI[i+1][j]*(1/(2*dz))*(omega[i+1][j+1]-omega[i+1][j-1]) - PHI[i-1][j]*(1/(2*dz))*(omega[i-1][j+1]-omega[i-1][j-1])) - (1/(2*dz))*(PHI[i][j+1]*(1/(2*dx))*(omega[i+1][j+1]-omega[i-1][j+1]) - PHI[i][j-1]*(1/(2*dx))*(omega[i+1][j-1]-omega[i-1][j-1]))    
    return (J_2)


def J_3_func (PHI, omega):
    J_3 =  np.zeros((Nx+2, Nz+2), dtype = float)
    for i in range(2,Nx):
        for j in range(2,Nz):
            J_3[i][j] = (1/(2*dz))*(omega[i][j+1]*(1/(2*dx))*(PHI[i+1][j+1]-PHI[i-1][j+1]) - omega[i][j-1]*(1/(2*dx))*(PHI[i+1][j-1] - PHI[i-1][j-1])) - (1/(2*dx))*(omega[i+1][j]*(1/(2*dz))*(PHI[i+1][j+1]-PHI[i+1][j-1]) - omega[i-1][j]*(1/(2*dz))*(PHI[i-1][j+1]-PHI[i-1][j-1]))
    return (J_3)

def J_final (PHI,omega):
    return ((1/3)*(J_3_func(PHI,omega)+J_2_func(PHI,omega)+J_1_func(PHI,omega)))



def MaxCourant(PHI, dt = 0.5,dx = 10 ): 
    absmaxU = np.amax(np.absolute(U_func(PHI)[2:-2,2:-2]))
    #print("Abs max U is", absmaxU)
    absmaxV = np.amax(np.absolute(V_func(PHI)[2:-2,2:-2]))
    #print("Abs max V is", absmaxV)
    return (absmaxU*dt/dx + absmaxV*dt/dx)

#IMPORTANT, RANGE (ghost nodes + walls ignored)
def totalKE (PHI):
    total = 0
    U = np.copy(U_func(PHI))
    V = np.copy(V_func(PHI))
    for i in range(2,Nx):
        for j in range(2,Nz):
            total += 0.5*(U[i][j]**2 + V[i][j]**2)
    return (total)

#IMPORTANT, omegaICITY AT WALL???????
def totalTemp (omega):
    total = 0
    for i in range(2,Nx):
        for j in range(2,Nz):
            total += (omega[i][j])
    return (total)

#NUMERICAL omegaICITY
def omega_NUM_func (PHI):
    omega_NUM = np.zeros((Nx+2, Nz+2), dtype = float)
    for i in range(1, Nx+1):
        for j in range(1,Nz+1):
            omega_NUM[i][j] = (1/dx**2)*(PHI[i-1][j] -2*PHI[i][j] + PHI[i+1][j])+(1/dz**2)*(PHI[i][j-1]-2*PHI[i][j]+ PHI[i][j+1])
    return (omega_NUM)
        

#******************************************************************************#
#***********************Initialize THETA    ***********************************#
#******************************************************************************#
theta_int = np.zeros((Nx+2, Nz+2), dtype = float)
theta = np.zeros((Nx+2, Nz+2), dtype = float)
omega = np.zeros((Nx+2, Nz+2), dtype = float)

for i in range(1,Nx+1): #201 physical node
    for j in range(2,Nz):
        if ( math.sqrt((dx*(i-1) - x_0)**2 + (dz*(j-1) - z_0)**2) <= r_0 ):
            theta_int[i][j] = theta_0 + d_theta
        else: 
            theta_int[i][j] = theta_0
            
theta = np.copy(theta_int)       

        
"""
#THETA BOUNDARY CONDITION

for j in range(0,Nz+2):
    theta_int[Nx+1][j] = 0
    theta_int[0][j] = 0
    theta_int[1][j] = theta_int[Nx][j]
    
for i in range(0,Nx+2):
    theta_int[i][Nz+1] = theta_int[i][2]
    theta_int[i][0] = theta_int[i][Nz-1]
    theta_int[i][1] = theta_int[i][Nz]
    
    

#OMEGA Boundary Condition

for j in range(0,Nz+2):
    omega[Nx+1][j] = 0
    omega[0][j] = 0
    omega[1][j] = omega[Nx][j]
for i in range(0,Nx+2):
    omega[i][Nz+1] = 0
    omega[i][0] = 0
    omega[i][1] = omega[i][Nz]
"""

print("finish theta field initialize")

print("Initial total temperature:",totalTemp(theta_int))
#not outputting the WALL *ghost nodes + 1 layer boundary
plt.figure(12109)
x = np.linspace(0,2000,199) 
y = np.linspace(0,2000,199)
X,Y = np.meshgrid(x,y)
plt.contourf(X,Y, np.transpose(theta_int[2:-2,2:-2]))
#plt.imshow( np.transpose(theta_int[1:-1,1:-1]), interpolation='none', extent=[0,2000,0,2000])
plt.title("Initial Theta Field")
plt.colorbar()
plt.xlabel("X Domain [m]")
plt.ylabel("Y Domain [m]")
#plt.savefig("INT,UNNORM-W-run_B.svg")




#******************************************************************************#
#*********************************** NORMALIZATION ****************************#
#******************************************************************************#

omega_Avg = twoDSum(omega[1:-1,1:-1])/(omega[1:-1,1:-1].size)

print("mean omega = ", omega_Avg)

"""
plt.figure(1)
plt.imshow( np.transpose(omega[1:-1,1:-1]), interpolation='none', extent=[0,100,0,100])
plt.colorbar()

plt.xlabel("X Domain [m]")
plt.ylabel("Y Domain [m]")
"""
#plt.savefig("INT,NORM-omega-run_B.svg")
#omega = np.transpose(omega_R)


#******************************************************************************#
#*****************POISSON ADVANCING LISTS**************************************#
#******************************************************************************#

PHI = np.zeros((Nx+2, Nz+2), dtype = float)
Residual = np.zeros((Nx+2, Nz+2), dtype = float)


#******************************************************************************#
#************************************ALPHA*************************************#
#******************************************************************************#


#F-PLANE APPROXIMATION
sigma = (1/(1+(dx/dz)**2))*(math.cos(math.pi/Nx) + math.cos(math.pi/Nz)*(dx/dz)**2)
alpha = 2/(1+math.sqrt(1-sigma**2))
print("sigma =", sigma)    
print("alpha =", alpha)

timeStep = []
EPScounter = []

#******************************************************************************#
#***********************ADVANCING LIST (JACOBIANS)*****************************#
#******************************************************************************#
J_older_omega = np.zeros((Nx+2, Nz+2), dtype = float) #n-2
J_old_omega = np.zeros((Nx+2, Nz+2), dtype = float)   #n-1
J_cur_omega = np.zeros((Nx+2, Nz+2), dtype = float)   #n
J_new_omega = np.zeros((Nx+2, Nz+2), dtype = float)   #n+1

J_older_theta = np.zeros((Nx+2, Nz+2), dtype = float) #n-2
J_old_theta = np.zeros((Nx+2, Nz+2), dtype = float)   #n-1
J_cur_theta = np.zeros((Nx+2, Nz+2), dtype = float)   #n
J_new_theta = np.zeros((Nx+2, Nz+2), dtype = float)   #n+1

theta_older = np.zeros((Nx+2, Nz+2), dtype = float) #n-2
theta_old= np.zeros((Nx+2, Nz+2), dtype = float)   #n-1
theta_cur = np.zeros((Nx+2, Nz+2), dtype = float)#n
theta_new = np.zeros((Nx+2, Nz+2), dtype = float)



omega_older = np.zeros((Nx+2, Nz+2), dtype = float) #n-2
omega_old= np.zeros((Nx+2, Nz+2), dtype = float)   #n-1
omega_cur = np.zeros((Nx+2, Nz+2), dtype = float)#n
omega_new = np.zeros((Nx+2, Nz+2), dtype = float)
#omega_new = np.zeros((Nx+2, Nz+2), dtype = float)


#******************************************************************************#
#*****************************FIRST TIME STEP**********************************#
#******************************************************************************#

#####POISSON SOLVER######

#ORDER is ******* IMPORTANT
#PHI = np.zeros((Nx+2, Nz+2), dtype = float)

#reset PSI to 0
PHI = np.zeros((Nx+2, Nz+2), dtype = float)

#Jacobian
J_older_omega = np.copy(J_final(PHI,omega))
J_older_theta = np.copy(J_final(PHI,theta))

theta_older = np.copy(theta_int)

for i in range(1,Nx+1):
    for j in range(1,Nz+1):
        omega_older[i][j] = omega_older[i][j] - J_older_omega[i][j]*dt - dt*(g/theta_0)*(theta_older[i+1][j] - theta_older[i-1][j])/(2*dx)
        omega_older[i][j] = omega_older[i][j] + Kx*(dt/(dx**2))*(omega_older[i+1][j] - 2*omega_older[i][j] + omega_older[i-1][j]) + Kz*(dt/(dz**2))*(omega_older[i][j+1] - 2*omega_older[i][j] + omega_older[i][j-1])
        theta_older[i][j] = theta_older[i][j] - (J_older_theta[i][j])*dt
        theta_older[i][j] = theta_older[i][j] + Kx*(dt/(dx**2))*(theta_older[i+1][j] - 2*theta_older[i][j] + theta_older[i-1][j]) + Kz*(dt/(dz**2))*(theta_older[i][j+1] - 2*theta_older[i][j] + theta_older[i][j-1])

omega = np.copy(omega_older)
theta = np.copy(theta_older)
#Boundary Condition
for j in range(0,Nz+2):
    omega[0][j] = omega[Nx-1][j]
    omega[Nx+1][j] = omega[2][j]
    omega[1][j] = omega[Nx][j]

    #VORT[1][j] = VORT[Nx][j]
for i in range(0,Nx+2):
    omega[i][Nz+1] = 0
    omega[i][0] = 0
    omega[i][1] = 0
    omega[i][Nz] = 0


      #Boundary Condition
for j in range(0,Nz+2):
    theta[0][j] = theta[Nx-1][j]
    theta[Nx+1][j] = theta[2][j]
    theta[1][j] = theta[Nx][j]
    #VORT[1][j] = VORT[Nx][j]
for i in range(0,Nx+2):
    theta[i][Nz+1] = theta_0
    theta[i][0] =  theta_0
    theta[i][1] =  theta_0
    theta[i][Nz] =  theta_0

omega_older = np.copy(omega)
theta_older = np.copy(theta)

for i in range(1,Nx+1):
    for j in range(2,Nz):
        Residual[i][j] = (1/dx**2)*(PHI[i-1][j] -2*PHI[i][j] + PHI[i+1][j]) + (1/dz**2)*(PHI[i][j-1] -2*PHI[i][j] + PHI[i][j+1]) - omega[i][j]

EPS = np.amax(np.absolute(Residual))/((2*((1/dx**2)+(1/dz**2))*twoDSum(np.absolute(PHI[1:-1,2:-2])))+ np.amax(np.absolute(omega[1:-1,2:-2])))
# = 1.0
print('This is EPS at 1 EF = ',EPS )


maincounter = 0

while (EPS > tol):
    for j in range(2,Nz):  
        if (j == Nz):  #(never touches)
            for z in range(0,Nx+2): 
                PHI[z][Nz+1] = PHI[z][2]
        #update finish

        for i in range(1,Nx+1): 
            #PERIODIC IN X   ######### (now never runs)
            if (i == Nx):
                PHI[Nx+1][j] = PHI[2][j]

            Residual[i][j] = (1/dx**2)*(PHI[i-1][j] -2*PHI[i][j] + PHI[i+1][j]) + (1/dz**2)*(PHI[i][j-1] -2*PHI[i][j] + PHI[i][j+1]) - omega[i][j]
            PHI[i][j] = (PHI[i][j] + (alpha*Residual[i][j])/(2*((1/dx**2)+(1/dz**2))))
            #PHI[i][j] = -PHI[i][j]

        #PERIODIC IN X  #############
        PHI[0][j] = PHI[Nx-1][j]
        PHI[1][j] = PHI[Nx][j] #make sure left right equal


    #update finish
    for t in range(0,Nx+2):
        PHI[t][0] = 0
        PHI[t][1] = 0
        PHI[t][Nz] = 0
        PHI[t][Nz+1] = 0

    EPS = np.amax(np.absolute(Residual))/((2*((1/dx**2)+(1/dz**2))*twoDSum(np.absolute(PHI[1:-1,2:-2])))+ np.amax(np.absolute(omega[1:-1,2:-2])))
    #print('EPS is', EPS, '  MainCounter =', maincounter, "Max of abs Value = "  , np.amax(np.absolute(Residual[2:-2,2:-2])))
    #print(" Index of max abs = ", np.unravel_index(tempResidual.argmax(), tempResidual.shape))   
    maincounter +=1

    #print(maincounter, EPS)
    
#**********make PHI negative**********#
for i in range(0,Nx+2):
    for j in range(0,Nz+2):
        PHI[i][j] = (-1)*PHI[i][j]        
#*************************************#  
print("1st EF DONE")
print("Total Temp:",totalTemp(theta_older))

#******************************************************************************#
#*****************************SECOND TIME STEP**********************************#
#******************************************************************************#

J_old_omega = np.copy(J_final(PHI,omega_older))
J_old_theta = np.copy(J_final(PHI,theta_older))


#reset PSI to 0
PHI = np.zeros((Nx+2, Nz+2), dtype = float)







for i in range(1,Nx+1):
    for j in range(1,Nz+1):
        omega_old[i][j] = omega_older[i][j] - J_old_omega[i][j]*dt - dt*(g/theta_0)*(theta_older[i+1][j] - theta_older[i-1][j])/(2*dx)
        omega_old[i][j] = omega_old[i][j] + Kx*(dt/(dx**2))*(omega_older[i+1][j] - 2*omega_older[i][j] + omega_older[i-1][j]) + Kz*(dt/(dz**2))*(omega_older[i][j+1] - 2*omega_older[i][j] + omega_older[i][j-1])
        theta_old[i][j] = theta_older[i][j] - (J_old_theta[i][j])*dt
        theta_old[i][j] = theta_old[i][j] + Kx*(dt/(dx**2))*(theta_older[i+1][j] - 2*theta_older[i][j] + theta_older[i-1][j]) + Kz*(dt/(dz**2))*(theta_older[i][j+1] - 2*theta_older[i][j] + theta_older[i][j-1])

omega = np.copy(omega_old)
theta = np.copy(theta_old)
#Boundary Condition
for j in range(0,Nz+2):
    omega[0][j] = omega[Nx-1][j]
    omega[Nx+1][j] = omega[2][j]
    omega[1][j] = omega[Nx][j]

    #VORT[1][j] = VORT[Nx][j]
for i in range(0,Nx+2):
    omega[i][Nz+1] = 0
    omega[i][0] = 0
    omega[i][1] = 0
    omega[i][Nz] = 0


      #Boundary Condition
for j in range(0,Nz+2):
    theta[0][j] = theta[Nx-1][j]
    theta[Nx+1][j] = theta[2][j]
    theta[1][j] = theta[Nx][j]
    #VORT[1][j] = VORT[Nx][j]
for i in range(0,Nx+2):
    theta[i][Nz+1] = theta_0
    theta[i][0] =  theta_0
    theta[i][1] =  theta_0
    theta[i][Nz] =  theta_0

omega_old = np.copy(omega)
theta_old = np.copy(theta)


for i in range(1,Nx+1):
    for j in range(2,Nz):
        Residual[i][j] = (1/dx**2)*(PHI[i-1][j] -2*PHI[i][j] + PHI[i+1][j]) + (1/dz**2)*(PHI[i][j-1] -2*PHI[i][j] + PHI[i][j+1]) - omega[i][j]

EPS = np.amax(np.absolute(Residual))/((2*((1/dx**2)+(1/dz**2))*twoDSum(np.absolute(PHI[1:-1,2:-2])))+ np.amax(np.absolute(omega[1:-1,2:-2])))
# = 1.0
print('This is EPS at 1 EF = ',EPS )

maincounter = 0

while (EPS > tol):
    for j in range(2,Nz):  
        if (j == Nz):  #(never touches)
            for z in range(0,Nx+2): 
                PHI[z][Nz+1] = PHI[z][2]
        #update finish

        for i in range(1,Nx+1): 
            #PERIODIC IN X   ######### (now never runs)
            if (i == Nx):
                PHI[Nx+1][j] = PHI[2][j]

            Residual[i][j] = (1/dx**2)*(PHI[i-1][j] -2*PHI[i][j] + PHI[i+1][j]) + (1/dz**2)*(PHI[i][j-1] -2*PHI[i][j] + PHI[i][j+1]) - omega[i][j]
            PHI[i][j] = (PHI[i][j] + (alpha*Residual[i][j])/(2*((1/dx**2)+(1/dz**2))))
            #PHI[i][j] = -PHI[i][j]

        #PERIODIC IN X  #############
        PHI[0][j] = PHI[Nx-1][j]
        PHI[1][j] = PHI[Nx][j] #make sure left right equal


    #update finish
    for t in range(0,Nx+2):
        PHI[t][0] = 0
        PHI[t][1] = 0
        PHI[t][Nz] = 0
        PHI[t][Nz+1] = 0

    EPS = np.amax(np.absolute(Residual))/((2*((1/dx**2)+(1/dz**2))*twoDSum(np.absolute(PHI[1:-1,2:-2])))+ np.amax(np.absolute(omega[1:-1,2:-2])))
    #print('EPS is', EPS, '  MainCounter =', maincounter, "Max of abs Value = "  , np.amax(np.absolute(Residual[2:-2,2:-2])))
    #print(" Index of max abs = ", np.unravel_index(tempResidual.argmax(), tempResidual.shape))   
    maincounter +=1

    #print(maincounter, EPS)
    
    
#**********make PHI negative**********#
for i in range(0,Nx+2):
    for j in range(0,Nz+2):
        PHI[i][j] = (-1)*PHI[i][j]        
#*************************************#  

print("2nd EF DONE")
print("Total Temp:",totalTemp(theta_old))





#******************************************************************************#
#***************************** MAIN TIME LOOP *********************************#
#******************************************************************************#

for T in range(2,Nt):
    
    J_cur_omega = np.copy(J_final(PHI,omega_old))
    J_cur_theta = np.copy(J_final(PHI,theta_old))


    #reset PSI to 0
    PHI = np.zeros((Nx+2, Nz+2), dtype = float)



    for i in range(1,Nx+1):
        for j in range(1,Nz+1):
            omega_cur[i][j] = omega_old[i][j] + (23/12)*(-(J_cur_omega[i][j])*dt - dt*(g/theta_0)*(theta_old[i+1][j]-theta_old[i-1][j])/(2*dx)) - (16/12)*((-J_old_omega[i][j])*dt - (g/theta_0)*dt*(theta_old[i+1][j] - theta_old[i-1][j])/(2*dx)) + (5/12)*((-J_older_omega[i][j])*dt - (g/theta_0)*dt*(theta_older[i+1][j] - theta_older[i-1][j])/(2*dx))  
            omega_cur[i][j] = omega_cur[i][j] + Kx*(2*dt/(dx**2))*(omega_older[i+1][j] - 2*omega_older[i][j] + omega_older[i-1][j]) + Kz*(2*dt/(dz**2))*(omega_older[i][j+1] - 2*omega_older[i][j] + omega_older[i][j-1])
            theta_cur[i][j] = theta_old[i][j] + (23/12)*(-J_cur_theta[i][j])*dt + (16/12)*dt*(J_old_theta[i][j]) - (5/12)*(J_older_theta[i][j])*dt
            theta_cur[i][j] = theta_cur[i][j] + Kx*(2*dt/(dx**2))*(theta_older[i+1][j] - 2*theta_older[i][j] + theta_older[i-1][j]) + Kz*(2*dt/(dz**2))*(theta_older[i][j+1] - 2*theta_older[i][j] + theta_older[i][j-1])

    omega = np.copy(omega_cur)
    theta = np.copy(theta_cur)

    #Boundary Condition
    for j in range(0,Nz+2):
        omega[0][j] = omega[Nx-1][j]
        omega[Nx+1][j] = omega[2][j]
        omega[1][j] = omega[Nx][j]

        #VORT[1][j] = VORT[Nx][j]
    for i in range(0,Nx+2):
        omega[i][Nz+1] = 0
        omega[i][0] = 0
        omega[i][1] = 0
        omega[i][Nz] = 0


          #Boundary Condition
    for j in range(0,Nz+2):
        theta[0][j] = theta[Nx-1][j]
        theta[Nx+1][j] = theta[2][j]
        theta[1][j] = theta[Nx][j]
        #VORT[1][j] = VORT[Nx][j]
    for i in range(0,Nx+2):
        theta[i][Nz+1] = theta_0
        theta[i][0] =  theta_0
        theta[i][1] =  theta_0
        theta[i][Nz] =  theta_0

    omega_cur = np.copy(omega)
    theta_cur = np.copy(theta)


    for i in range(1,Nx+1):
        for j in range(2,Nz):
            Residual[i][j] = (1/dx**2)*(PHI[i-1][j] -2*PHI[i][j] + PHI[i+1][j]) + (1/dz**2)*(PHI[i][j-1] -2*PHI[i][j] + PHI[i][j+1]) - omega[i][j]

    EPS = np.amax(np.absolute(Residual))/((2*((1/dx**2)+(1/dz**2))*twoDSum(np.absolute(PHI[1:-1,2:-2])))+ np.amax(np.absolute(omega[1:-1,2:-2])))
    # = 1.0
    

    maincounter = 0

    while (EPS > tol):
        for j in range(2,Nz):  
            if (j == Nz):  #(never touches)
                for z in range(0,Nx+2): 
                    PHI[z][Nz+1] = PHI[z][2]
            #update finish

            for i in range(1,Nx+1): 
                #PERIODIC IN X   ######### (now never runs)
                if (i == Nx):
                    PHI[Nx+1][j] = PHI[2][j]

                Residual[i][j] = (1/dx**2)*(PHI[i-1][j] -2*PHI[i][j] + PHI[i+1][j]) + (1/dz**2)*(PHI[i][j-1] -2*PHI[i][j] + PHI[i][j+1]) - omega[i][j]
                PHI[i][j] = (PHI[i][j] + (alpha*Residual[i][j])/(2*((1/dx**2)+(1/dz**2))))
                #PHI[i][j] = -PHI[i][j]

            #PERIODIC IN X  #############
            PHI[0][j] = PHI[Nx-1][j]
            PHI[1][j] = PHI[Nx][j] #make sure left right equal


        #update finish
        for t in range(0,Nx+2):
            PHI[t][0] = 0
            PHI[t][1] = 0
            PHI[t][Nz] = 0
            PHI[t][Nz+1] = 0

        EPS = np.amax(np.absolute(Residual))/((2*((1/dx**2)+(1/dz**2))*twoDSum(np.absolute(PHI[1:-1,2:-2])))+ np.amax(np.absolute(omega[1:-1,2:-2])))
        #print('EPS is', EPS, '  MainCounter =', maincounter, "Max of abs Value = "  , np.amax(np.absolute(Residual[2:-2,2:-2])))
        #print(" Index of max abs = ", np.unravel_index(tempResidual.argmax(), tempResidual.shape))   
        maincounter +=1
        #print(maincounter, EPS)



    #**********make PHI negative**********#
    for i in range(0,Nx+2):
        for j in range(0,Nz+2):
            PHI[i][j] = (-1)*PHI[i][j]        
    #*************************************#  

    
    #Boundary Condition
    for j in range(0,Nz+2):
        PHI[0][j] = PHI[Nx-1][j]
        PHI[Nx+1][j] = PHI[2][j]
        PHI[1][j] = PHI[Nx][j]

        #VORT[1][j] = VORT[Nx][j]
    for i in range(0,Nx+2):
        PHI[i][Nz+1] = 0
        PHI[i][0] = 0
        PHI[i][1] = 0
        PHI[i][Nz] = 0

        
        
        
        
    for i in range(0,Nx+2):
        for j in range(0,Nz+2):
            if (theta_cur[i][j] <= theta_0):
                theta_cur[i][j] = theta_0
    #shifting: 
                         
    J_older_omega = np.copy(J_old_omega)
    J_old_omega = np.copy(J_cur_omega)
    
    J_older_theta = np.copy(J_old_theta)
    J_old_theta = np.copy(J_cur_theta)


    theta_older = np.copy(theta_old)
    theta_old = np.copy(theta_cur)

    omega_older = np.copy(omega_old)
    omega_old= np.copy(omega_cur)

   


    print("Total Temp:", totalTemp(theta_cur))
    print("SOR Iteration Counter =", maincounter )  
    print("Time Step =", T)
    #print("total Temp =", totalTemp(theta))
    if (T%50 == 0):        


        plt.figure(101231+T)
        plt.imshow( theta_cur[1:-1,1:-1].T ,origin ='lower', interpolation='none', extent=[0,Lx,0,Lz])
        plt.colorbar()
        plt.title("Theta" + str(T))
        plt.xlabel("X Domain [m]")
        plt.ylabel("Y Domain [m]")
        #plt.savefig("Theta at"+str(T)".svg")




```
